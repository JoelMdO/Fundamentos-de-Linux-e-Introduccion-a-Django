## Ciclos For

### test_templates/views.py

```python
def testView(request):
    
    my_list = [
        "mouse", "laptop", "teclado", "audifonos", "celular"
    ]

    context={
        "view_title": "My Titulo",
        "my_number": 675,
        "my_number_2":2000,
        "today": datetime.now().today()
    }

    template = test_template/test_view.html

    return render(request, template, context)
```

### templates/test_templates/test_view.html

```python
Django docs : templates/filters


<style>
 .bold {
    font-weight: "bold";
 }
</style>

<ul>
    {% for obj in my_list %}
    <li {% if forloop.first %} class="bold" {% endif %}>
        {{forloop.counter}}
    </li>
    {% endfor %}
</ul>

## Pares
<ul>
    {% for obj in my_list %}
    <li {% if forloop.counter|divisibleBy:2 %} class="bold" {% endif %}>
        {{forloop.counter}}{{obj}}
    </li>
    {% endfor %}
</ul>

## Impares
<ul>
    {% for obj in my_list %}
    <li {% if forloop.counter|add:1|divisibleBy:2 %} class="bold" {% endif %}>
        {{forloop.counter}}{{obj}}
    </li>
    {% endfor %}
</ul>

## Al tercero en la lista aplica bold
<ul>
    {% for obj in my_list %}
    <li class='{% cycle "" "" "bold" %}' {% endif %}>
        {{forloop.counter}}{{obj}}
    </li>
    {% endfor %}
</ul>
```
