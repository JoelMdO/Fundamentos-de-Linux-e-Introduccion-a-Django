### Crear ListView protegido por login

products/views.py

```python
class ProtectedProductListView(LoginRequiredMixin, ListView):
    model = Product

    def get_queryset(self) -> QuerySet[Any]:
        return super().get_queryset().filter(user=self.request.user)
```
