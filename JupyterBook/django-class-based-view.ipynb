### Django Class-based Views


- Agregamos la vista en products/views.py
-   Método “get_queryset” para que aparezcan solo los productos del usuario que hace el request.

```python
from django.contrib.auth.mixins import LoginRequiredMixin
from django.views.generic import ListView

class ProtectedProductListView(LoginRequiredMixin, ListView):
    model = Product
    template = "products/protected_listview.html"
    
    def get_queryset(self) -> QuerySet[Any]:
        return super().get_queryset().filter(user=self.request.user)
```
