### Agregar 500 objetos de Prueba de manera automatizada

### Pasos
- Ingresar al contenedor

```python
docker exec -it hellodjango-web-1 bash
```

- Ingresar al shell

```python
python@86e28313f7ad:/app/src$ python manage.py shell
```

- Dar el comando de ejecucion como una sola linea en terminal

```python
from ecommerce.models import ProductModel

for i in range(500):
    obj, created = ProductModel.objects.get_or_create(
        title=f"Producto {i}",
        defaults={"price": i*100 + 99.99}
    )
    if created and (i+1) % 50 == 0:
        print("Created:", i+1, flush=True)
```

- Exit the shell

```python
exit()
```
