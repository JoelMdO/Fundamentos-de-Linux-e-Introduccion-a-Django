# Programa para imprimir el numero de veces que cada palabra se repite

```python
texto = input("Escribe una frase: ")
conteo = {}

# comience su código aquí:
def countWords (texto):
    
    textoArray = texto.split()

    for word in textoArray:
        if word in conteo:
          conteo[word] += 1
        else:
          conteo[word] = 1

    for palabra, total in conteo.items():
      print(f"La palabra '{palabra}', se repite '{total}' instancia(s)")

    return conteo

result = countWords(texto)
```

Las reglas se muestran a continuacion:

```python
pago_total = 0
cadena_horas_laboradas = input("Ingrese el número de horas laboradas en una semana: ")
cadena_tarifa_por_hora = input("Ingrese su tarifa de pago por hora: ")

# comience su código aquí:
#El programa deberá calcular el total a pagar al trabajador de acuerdo a las siguientes reglas:
def horas40To50(hours, extraHours, rate):
        subTotal = 0
        newRate = rate + (rate * 0.1)
        subTotal = (extraHours * newRate)
        return subTotal + (40 * rate)

def totalPayment(cadena_horas_laboradas, cadena_tarifa_por_hora):
# d) En caso de que alguno de los valores capturados inicialmente (número de horas trabajadas por semana o tarifa de pago por hora) sea errónea (ya sea debido a la captura de valores negativos o de texto), deberá imprimirse un mensaje que diga “Parámetros incorrectos”.
    try :
        horas = int(cadena_horas_laboradas) 
        rate = int(cadena_tarifa_por_hora)
        print(f"horas, '{horas}'")
        print(f"rate, '{rate}'")
        print(f"horastype, '{type(horas)}'")
        if horas <= 0 or rate <= 0:
            print("Parámetros incorrectos")
            return
# a) Si el número de horas trabajadas es menor o igual a 40 el monto a pagar es el número de horas multiplicado por su tarifa de pago por hora.
        elif horas <= 40 :
            pago_total = horas * rate
            print(f"pagototal {pago_total}")   
# b) Si el número de horas trabajadas es mayor a 40 pero menor o igual a 50, el monto a pagar es el mismo que en el inciso a) más la diferencia en horas por arriba de 40 a una tasa por hora 10% más alta de la tarifa básica.
        elif horas > 40 and horas <= 50:
            extraHours = horas - 40
            pago_total = horas40To50(horas, extraHours, rate)
# c) Si el número de horas trabajadas es mayor a 50, el monto a pagar es el mismo que en el inciso b) más la diferencia en horas por arriba de 50 a una tasa por hora 20% más alta de la tarifa básica.
        elif horas > 50 :
            newRateAbove50 = rate + (rate * 0.2) 
            extraHoursAbove50 = horas - 50
            print(f"extraHoursAbove50 {extraHoursAbove50}")
            extraHoursUpTo40 = horas - (40 + extraHoursAbove50)
            print(f"extraHoursUpTo40 {extraHoursUpTo40}")
            subTotal40To50 = horas40To50(horas, extraHoursUpTo40, rate)
            print(f"subTotal40To50, {subTotal40To50}")
            subTotalAbove50 = (extraHoursAbove50 * newRateAbove50)
            print(f"subTotalAbove50, {subTotalAbove50}")
            pago_total = subTotal40To50 + subTotalAbove50

#El valor final debe almacenarse en la variable pago_total y debe redondearse a 1 decimal.
        return round(pago_total,1)
    except :
        print("Parámetros incorrectos")

pago_total = totalPayment(cadena_horas_laboradas, cadena_tarifa_por_hora)

print(f"Número de horas trabajadas: {cadena_horas_laboradas}, Tarifa {cadena_tarifa_por_hora}, Monto a pagar: {pago_total}")
```

# Programa de factorial de un numero


```python
cadena = input("Dame un entero positivo: ")

# Puedes usar la estructura "try/except" que aprendimos en el módulo 2.
# comience su código aquí:

#Desarrolle un programa en Python que mediante el uso de ciclos calcule el factorial del número que indique el usuario. En caso de que el valor proporcionado sea inválido (por ejemplo, que sea negativo o fraccionado), deberá devolver un mensaje de "Error de captura". Pruebe su código calculando los factoriales de 4, 5 y 6 respectivamente así como el de 7.5

def factorialCalc(numero):
	fact = 1
	
	try:
		n = int(numero)
		if n < 0 or n % 1 !=0:
			print(f"Error de captura")

		while n >= 1:
			print(f"n", n)
			fact = fact * n 
			print(f"factorial", fact)
			n = n - 1
			print(f"nueva n", n)
		return fact
	except:
			print(f"Error de captura")

factorial = factorialCalc(cadena)
print(f"factorial de {cadena}: {factorial}")
#Almacena el resultado en la varible factorial.
#El factorial de un número entero positivo n, representado por n!, es la multiplicación de n por sus antecesores mayores o iguales a 1. Por ejemplo: 4! = 4x3x2x1 = 24:
```

# Progama de numero Primo

```python
def prueba_primo(numero):
    print(f"numero {numero}")
    if numero <= 1:
        return False
 
    for i in range(2,numero):
        if numero % i == 0:
            return False

    return True

numero = 7
print(f"Es {numero} primo? {prueba_primo(numero)}")
```

# Programa uso de Clases para Conversion de Numeros Romanos

```python
class Conversion:

    def __init__(self):
     self.values = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}

    def romano_entero(self, numero):
        total = 0
        valor_previo = 0

        for letra in reversed(numero):
            valor = self.values[letra]
            if valor < valor_previo:
                total -= valor
            else:
                total += valor
        
            valor_previo = valor
        return total

numero = Conversion()
print(f"La Conversion de {numero} es: {numero.romano_entero('MMMCMLXXXVI')}")
```

# Programa Conversion de Grados a Centrigrados

```python
import numpy as np

def fahrenheit_a_centigrados(array):
	# escribe la lógica de la función aquí
    centigrados = np.zeros(len(array), dtype=float)
    for i in range(len(array)):
        farenheit = array[i]
        degreeC = (5*(farenheit-32)/9)
        centigrados[i] = np.round(degreeC, 2)
    return centigrados

array = np.array([0, 12, 45.21, 34, 99.91, 32])
print("Valores en grados Fahrenheit:")
print(array)
print("Valores en grados centígrados")
# inserte la última línea de la impresión a continuación 
print(fahrenheit_a_centigrados(array))
```

# Programa de Union de arreglos

```python
import numpy as np

def union_de_arreglos(array1, array2):
	# escribe la lógica de la función aquí
    if not isinstance(array1, np.ndarray) or not isinstance(array2, np.ndarray):
        return "Array is not numpy"
    else:
        newArray = np.union1d(array1, array2)
        return newArray

array1 = np.array([1,2,3,4,5])
array2 = np.array([2,4,6,7,8])
print(union_de_arreglos(array1, array2))
# imprime, a continuación, el resultado de la función
```
