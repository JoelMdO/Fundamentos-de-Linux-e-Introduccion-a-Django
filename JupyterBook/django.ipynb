## Estructura del Proyecto en Django
### Archivos de Proyecto
- manage.py : Es un archivo que contiene muchas funciones que te permiten interactuar con Django.
- config/__ init __.py: Es un archivo vacio que le informa a python que este directorio/carpeta deberia ser considerado como un paquete de Python.
- config/settings.py: Configuracion o ajustes para el proyecto.
- config/urls.py: Las declaraciones de los urls para el proyecto, es una tabla de contenidos con los endpoints del sitio web.
- config/asgi.py: Un punto de entrada para servidores web compatibles con ASGI para desplegar el proyecto.
- config/wsgi.py: Un punto de entrada para servidores web compatibles con WSGI para desplegar el proyecto.

### Archivos de Aplicacion
- admin.py: Registrar tus modelos relacionados con el django admin.
- apps.py: Contiene el codigo de la configuracion de la app.
- models.py: Define los modelos que son tablas para definir la base de datos y sus relaciones.
- tests.py: Escribir tests para la app.
- views.py: Definir la informacion que sera mandada a los templates del frontend.

## Primeros cambios en el proyecto.

```python
import os
from typing import Dict, Any

from django import get_version
from django.conf import settings
from django.shortcuts import render
from django.http import HttpRequest


def home(request: HttpRequest):
    context: Dict[str, Any] = {
        "debug": settings.DEBUG,
        "django_ver": get_version() + "PROBANDO CAMBIOS",
        "python_ver": os.environ["PYTHON_VERSION"] + "MAS CAMBIOS",
    }

    return render(request, "pages/home.html", context)
```
